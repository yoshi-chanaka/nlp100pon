```python
# 2.1 明確な単語を選ぶ
def ExtractMorphemesInfo(wakati_result):

    morphemes_inSentence_list = []
    
    for line in wakati_result.split('\n'): # EOSを除く
        
        morpheme_info_dict = {}
        line_tab_splitted = line.split('\t')
        info_of_morpheme = line_tab_splitted[1].split(',')
        
        # 4.4 縦の線をまっすぐにする
        morpheme_info_dict['surface']   = line_tab_splitted[0]  # 表層形
        morpheme_info_dict['base']      = info_of_morpheme[6]   # 基本形
        morpheme_info_dict['pos']       = info_of_morpheme[0]   # 品詞
        morpheme_info_dict['pos1']      = info_of_morpheme[1]   # 品詞細分類1

        morphemes_inSentence_list.append(morpheme_info_dict)

    return morphemes_inSentence_list
```

```python
MORPHEMES_INFO_NEKO = []
wakati_result = ''
path = '../data/neko.txt.mecab'
with open(path, 'r', encoding='utf-8') as f:
    line = f.readline()
    while line:
        if line.strip() == 'EOS':
            MORPHEMES_INFO_NEKO.append(ExtractMorphemesInfo(wakati_result.strip()))
            wakati_result = ''
        else:
            wakati_result += line
        line = f.readline()
```

```python
noun_phrases_list = [] # 名詞句

# 4.7 コードを段落に分割する
for morphs_in_sentence in MORPHEMES_INFO_NEKO:
    
    # 文章の品詞（pos）と表層形（surface）のリストを取得
    pos_in_sentence, surface_in_sentence = [], []
    for morph_info in morphs_in_sentence:
        pos_in_sentence.append(morph_info['pos'])
        surface_in_sentence.append(morph_info['surface'])

    # AのB（A，Bは名詞）となっている部分を抽出
    for idx in range(len(pos_in_sentence)-2):
        if pos_in_sentence[idx: idx + 3] == ['名詞', '助詞', '名詞'] and \
            surface_in_sentence[idx + 1] == 'の':
            noun_phrases_list.append(''.join(surface_in_sentence[idx: idx + 3]))
```

```python
import numpy as np
import re

# 10章 無関係の下位問題（汎用コード）を抽出
def MaxLenConsecutivePos(pos_sequence, key_pos='名詞'):
    # リストからkey_posが連続して出現する回数の最大値を返す

    if key_pos not in pos_sequence:
        return 0
    pos_binary_labels = ''.join(['1' if pos == key_pos else '0' for pos in pos_sequence])    # 1100111101
    pos_binary_labels = re.sub(r'00+', '0', pos_binary_labels)                                # 110111101
    continuity_one_list = ''.join(pos_binary_labels).split('0')                             # 11, 1111, 1
    num_continuity_ones = [int(np.log2(int(n, 2) + 1)) for n in continuity_one_list if n != '']
    # 1111(str, 2進) => 15 (int) => 16 => log2(16) = 4

    return max(num_continuity_ones)
```

```python
print(MaxLenConsecutivePos(['名詞', '名詞', '動詞', '形容詞', '名詞', '名詞', '名詞', '名詞', '形容動詞'])) # 4
print(MaxLenConsecutivePos(['形容動詞'])) # 0
print(MaxLenConsecutivePos(['名詞', '名詞', '動詞', '形容詞', '名詞', '名詞', '名詞', '名詞', '名詞', '名詞', '名詞', '形容動詞'])) # 7
```

```python
# 名詞の連続出現回数と文章IDはここで保持
max_consecutive_pos_holder = {'max': -1, 'sentenceID': []}

for sentence_idx, morphs_in_sentence in enumerate(MORPHEMES_INFO_NEKO):
    pos_sequence = [morph['pos'] for morph in morphs_in_sentence]
    max_num = MaxLenConsecutivePos(pos_sequence)

    # 名詞の連続出現回数がholderに格納された回数よりも大きい場合
    if max_consecutive_pos_holder['max'] < max_num:
        max_consecutive_pos_holder['max'] = max_num
        max_consecutive_pos_holder['sentenceID'] = [sentence_idx]

    # 名詞の連続出現回数がholderに格納された回数と同じ
    elif max_consecutive_pos_holder['max'] == max_num:
        max_consecutive_pos_holder['sentenceID'].append(sentence_idx)
```
