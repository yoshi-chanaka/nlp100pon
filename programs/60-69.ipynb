```python
import numpy as np

def ComputeAcc(y_true, y_pred):
  if (len(y_true) == 0) or (len(y_true) != len(y_pred)):
    return False
  return np.sum(np.array(y_true) == np.array(y_pred)) / len(y_true)
```

```python
# 4.7 コードを「段落」に分割する
# 4.2 一貫性のある簡潔な改行位置：似ているコードは似ているように見せる
analogies_dict = {'semantic': {'true': [], 'pred': []},
                'syntactic': {'true': [], 'pred': []}}
analogy = 'semantic'

f = open('../data/questions-words_similarity.txt', 'r', encoding='utf-8')
line = f.readline()
while analogy == 'semantic':
    if line[0] == ':':
        kind_of_analogy = line[2:6]
        if kind_of_analogy == 'gram':
            analogy = 'syntactic'
    else:
        line_splitted = line.strip().split()
        analogies_dict[analogy]['true'].append(line_splitted[3])
        analogies_dict[analogy]['pred'].append(line_splitted[4])
    line = f.readline()

while line: # syntactic
    if line[0] == ':':
        pass
    else:
        line_splitted = line.strip().split()
        analogies_dict[analogy]['true'].append(line_splitted[3])
        analogies_dict[analogy]['pred'].append(line_splitted[4])
    line = f.readline()

f.close()
```

```python
print('意味的アナロジーの正解率:', ComputeAcc(analogies_dict['semantic']['true'], analogies_dict['semantic']['pred']))
print('文法的アナロジーの正解率:', ComputeAcc(analogies_dict['syntactic']['true'], analogies_dict['syntactic']['pred']))
```

```python
# 6.5 入出力のコーナーケースに実例を使う
"""
New Zealand => New_Zealand
Saint Vincent and the Grenadines => Saint_Vincent, Grenadines
United States of America => United_States, America
Congo (Congo-Brazzaville) => Congo, Congo_Brazzaville
みたいにしたい
"""

import re

countries = []
f = open('../data/countrylist.txt', 'r', encoding='utf-8')
line = f.readline().strip()
while line:
    line = line.replace(' the ', ' ')
    line = re.sub('[-\s]', '_', line)
    line = re.sub('\)', '', line)
    line = line.replace('_of_', '|').replace('_and_', '|').replace('_(', '|')
    countries += line.split('|')

    line = f.readline().strip()
f.close()

countries = list(set(countries))
```

取得した国名のうち，word2vecのvocabにも含まれるもののベクトルを分析対象とする．<br>
United_StatesとAmericaなど，同じ国だが別の表記であるものも存在するが，両方とも分析対象とする．

```python
type(model.key_to_index.keys())
```

```python
w2v_countries = {'country': [],
                'vectors': []}
w2v_vocab = set(list(model.key_to_index.keys()))
for countryname in countries:
    if countryname in w2v_vocab:
        w2v_countries['country'].append(countryname)
        w2v_countries['vectors'].append(model[countryname])

w2v_countries['vectors'] = np.array(w2v_countries['vectors'])
print(len(w2v_countries['country']), w2v_countries['vectors'].shape)
"""
196 (196, 300)
"""
```
