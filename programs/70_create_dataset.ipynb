```python
from gensim.models import KeyedVectors
path = './GoogleNews-vectors-negative300.bin'
model = KeyedVectors.load_word2vec_format(path, binary=True)
```

```python
import numpy as np

category2label = {'b': 0, 't': 1, 'e': 2, 'm': 3}
titles = {'train': [], 'valid': [], 'test': []}
y = {'train': [], 'valid': [], 'test': []}

for k in ['train', 'valid', 'test']:
  f = open('./data/' + k + '.txt', 'r')
  line = f.readline()
  while line:
    res = line.strip().split(',')
    titles[k].append(res[1])
    y[k].append(category2label[res[4]])
    line = f.readline()
  f.close()
```

```python
def compute_embedding(text):

  global model

  word_embeddings = []
  for w in text.split(' '):
    try:
      word_embeddings.append(model[w].tolist())
    except:
      pass
  
  if len(word_embeddings) == 0:
    print('error')
    return False
  else:
    word_embeddings = np.array(word_embeddings)

  return np.mean(word_embeddings, axis=0)
```

```python
X = {}
for k in ['train', 'valid', 'test']:
  X[k] = np.array([compute_embedding(t) for t in titles[k]])
  y[k] = np.array(y[k])
  print(f'{k}: {X[k].shape}, {y[k].shape}')
```

```python
data = {}
for k in ['train', 'valid', 'test']:
  data['X_' + k] = X[k]
  data['y_' + k] = y[k]
```

pickleファイルに保存

```python
import pickle

with open('./chap08_data.pickle', 'wb') as f:
    pickle.dump(data, f)
```
